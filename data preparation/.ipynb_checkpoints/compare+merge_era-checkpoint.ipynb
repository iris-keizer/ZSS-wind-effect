### Make figures for comparison

```python
def plot_timeseries(era5_data, erainterim_data, xstart = 1979, xend = 2021):
    """
    Function to make a plot of the wind stress data of ERA5 and ERA-interim
    
    """
    
    
    fig, axs = plt.subplots(4, 2, figsize=(10, 10))


    for i in range(4):


        ax = axs[i,0]
        ax.plot(era5_data.index, era5_data[era5_data.columns[2*i]].values)
        ax.plot(erainterim_data.index, erainterim_data[erainterim_data.columns[2*i]].values)
        ax.set_title(era5_data.columns[2*i][0], fontsize = 14)
        ax.set_xlabel('Time [yr]', fontsize = 12)
        ax.set_ylabel('Wind stress', fontsize = 12)
        ax.set_xlim(xstart, xend)
        ax.set_xticks(range(xstart+1, xend, 10))
        plt.tight_layout()


        ax = axs[i,1]
        if i == 3:
            fig.delaxes(axs[3,1])
        else:
            
            ax.plot(era5_data.index, era5_data[era5_data.columns[2*i]].values)
            ax.plot(erainterim_data.index, erainterim_data[erainterim_data.columns[2*i]].values)
            ax.set_title(erainterim_data.columns[2*i+1][0], fontsize = 14)
            ax.set_xlabel('Time [yr]', fontsize = 12)
            ax.set_ylabel('Wind stress', fontsize = 12)
            ax.set_xlim(xstart, xend)
            ax.set_xticks(range(xstart+1, xend, 10))
            plt.tight_layout()
    
    
    labels=['ERA5', 'ERA-interim']
    fig.legend(labels=labels, loc=(0.57, 0.1))
    
    plt.savefig(f'/Users/iriskeizer/Documents/Zeespiegelscenarios/figures/comparison ERA5 ERA-interim/comparison_{era5_data.columns[0][1][1]}_{xstart}_{xend}')
    
```

```python
plot_timeseries(windstress_era5[lst_u2], windstress_erainterim[lst_u2])
```

```python
plot_timeseries(windstress_era5[lst_u2], windstress_erainterim[lst_u2], 1999, 2021)
```

```python
plot_timeseries(windstress_era5[lst_v2], windstress_erainterim[lst_v2])
```

```python
plot_timeseries(windstress_era5[lst_v2], windstress_erainterim[lst_v2], 1999, 2021)
```

## Combine ERA-interim 1980-2017 and ERA5 2018-2020

```python
windstress_erainterim_begin = windstress_erainterim.loc[:2017]
windstress_era5_end = windstress_era5.loc[2018:]
```

```python
windstress = pd.concat([windstress_erainterim_begin, windstress_era5_end])
```

```python
path = '/Users/iriskeizer/Documents/Zeespiegelscenarios/data/wind stress'

windstress.to_csv(f'{path}/windstress.csv')
```
